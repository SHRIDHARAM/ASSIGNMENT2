```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
x1 = np.linspace(-4,4,100)
y1 = (-3*x1 + 9 )/4
plt.plot(x1, y1, '-b', label='3x+4y-9=0')

x2 = np.linspace(-4,4,100)
y2 = (-12*x2 + 3 )/5
plt.plot(x2, y2, '-b', label='12x+5y-3=0')

x3 = np.linspace(-4,4,100)
y3 = (-9*x3 + 12 )/7
plt.plot(x3, y3, '-g', label='9x+7y-12=0')

x4 = np.linspace(-4,4,100)
y4 = (7*x4 + 34 )/9
plt.plot(x4, y4, '-m', label='7x-9y+34=0')

plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend(loc='best')
plt.grid()
plt.xlim(-5, 5)
plt.ylim(-20, 20)
plt.show()
```
